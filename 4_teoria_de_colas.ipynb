```python
import numpy as np

# Parámetros del sistema
tiempo_llegada = 12  # tiempo medio entre llegadas (minutos)
tiempo_servicio = 4  # tiempo medio de servicio (minutos)

# Tasas
lambda_llegada = 1 / tiempo_llegada  # tasa de llegada
mu_servicio = 1 / tiempo_servicio    # tasa de servicio
rho = lambda_llegada / mu_servicio   # factor de utilización

# a) Tiempo medio en el sistema
W = 1 / (mu_servicio - lambda_llegada)
print(f"a) Tiempo medio en el sistema: {W:.2f} minutos")

# b) Número promedio de servidores ocupados
L_server = rho
print(f"b) Número promedio de servidores ocupados: {L_server:.4f}")

# c) Porcentaje de ocupación del servidor
ocupacion = rho * 100
print(f"c) Porcentaje de ocupación del servidor: {ocupacion:.2f}%")

# Verificación con las probabilidades dadas
probs = {0: 0.6667, 1: 0.2222, 2: 0.0741, 3: 0.0247, 4: 0.0082,
         5: 0.0027, 6: 0.0009, 7: 0.0003, 8: 0.0002}

# Calculamos el número promedio de clientes en el sistema usando las probabilidades
L = sum(i * prob for i, prob in probs.items())
print(f"Número promedio de clientes en el sistema: {L:.4f}")
```

```python
from math import factorial

def calcular_wq(lam, mu, s):
    rho = lam / (s * mu)
    if rho >= 1:
        return float('inf')

    a = lam / mu
    sum_term = sum([(a**n)/factorial(n) for n in range(s)])
    last_term = (a**s) / (factorial(s) * (1 - rho))
    P0 = 1 / (sum_term + last_term)

    Lq = ((a**s) * rho) / (factorial(s) * (1 - rho)**2) * P0
    Wq = Lq / lam
    return Wq

# Parámetros
lam = 0.75  # llegadas por minuto
mu = 0.2    # servicio por minuto
limite_wq = 0.5  # minutos

# Iterar para encontrar el mínimo s
for s in range(1, 20):
    wq = calcular_wq(lam, mu, s)
    print(f"s = {s}, Wq = {wq:.4f}")
    if wq <= limite_wq:
        print(f"✅ Se necesitan al menos {s} cajas (servidores). Wq = {wq:.4f} minutos")
        break
```
