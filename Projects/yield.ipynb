```python
import random
```

```python
def random_walker_clasico(pasos):
    x = 0
    for _ in range(pasos):
        x += random.choice([-1, 1])
    return x
```

```python
caminante_1 = random_walker_clasico(100)
print(caminante_1)
```

```python

# -----------------------
# Random Walker Generator
# -----------------------
def random_walker(pasos):
    x = 0
    for _ in range(pasos):
        x += random.choice([-1, 1, 2, -2])
        yield x
```

```python
walker = random_walker(2)
for _ in range(2):
    print(next(walker))
```

```python
print(walker)
```

```python
# imprimir el siguiente valor generado
for _ in range(4):
    print(next(walker))
```
